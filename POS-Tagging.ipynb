```python
import nltk
```

```python
from nltk.tokenize import word_tokenize
```

```python
from nltk import pos_tag
```

```python
nltk.download("all")
```

```python
nltk.download("averaged_perceptron_tagger")
```

```python
text = "the quick white rabbit jumps over the lazy browm dog "
```

```python
text
```

```python
words = word_tokenize(text, language = "english")
```

```python
words
```

```python
pos = pos_tag(words)
```

```python
pos
```

```python
nltk.help.upenn_tagset("NN")
```

```python
nltk.help.upenn_tagset("NNS")
```

```python
import time
```

```python
for i in pos :
    print(i)
    print(nltk.help.upenn_tagset(i[1]) )
    time.sleep(1)
```
