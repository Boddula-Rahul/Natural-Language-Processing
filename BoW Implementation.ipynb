```python
import pandas as pd
```

```python
import numpy as np
```

```python
sent = ["Python is Multi Purpose Programming Language ", 
         "I'm a Python Developer love to programming in Python ",
       "Python is Object Oriented Programming Language"]
```

```python
sent
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cvec = CountVectorizer()
```

```python
cvec
```

```python
sent
```

```python
cvec.fit(sent)
```

```python
cvec.get_feature_names_out()
```

```python
len(cvec.get_feature_names_out())
```

```python
cvec.vocabulary_
```

```python
cvec.transform(sent)
```

```python
ary = cvec.transform(sent).toarray()
```

```python
ary
```

```python
sent
```

```python
pd.DataFrame(ary, columns =  cvec.get_feature_names_out() )
```
